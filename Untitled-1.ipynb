# Analyze and Plot Time Series for Group 2 x497

This notebook reads the MATLAB `.mat` file for Group 2 x497, extracts the time series data, and plots it for analysis.

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# plot x and y velocity profiles for Group 2 x497
# Use scatter plots with small dots
def plt_velocity_profiles(t_time, x_vel, y_vel, name):
    # Restrict to 0-30 seconds
    t = np.array(t_time).flatten()
    mask = (t >= 0) & (t <= 30)
    t = t[mask]
    x_v = np.array(x_vel).flatten()[mask]
    y_v = np.array(y_vel).flatten()[mask]
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plt.scatter(t, x_v, label='X Velocity', color='b', s=5)
    plt.xlabel('Time (s)')
    plt.ylabel('X Velocity (m/s)')
    plt.xlim(0,30)
    plt.title('X Velocity Profile Over Time (Group 2 ' + name + ')')
    plt.legend()
    plt.grid()

    plt.subplot(2, 1, 2)
    plt.scatter(t, y_v, label='Y Velocity', color='r', s=5)
    plt.xlabel('Time (s)')
    plt.xlim(0,30)
    plt.ylabel('Y Velocity (m/s)')
    plt.title('Y Velocity Profile Over Time (Group 2 ' + name + ')')
    plt.legend()
    plt.grid()
    plt.tight_layout()
```

```python
def furier_analysis(t_time, x_vel, y_vel, name):
    # Define intervals
    intervals = [(0, 30), (30, 60), (60, 90), (90, 120)]
    fft_x_list = []
    fft_y_list = []
    freq_common = None
    for start, end in intervals:
        t = np.array(t_time).flatten()
        mask = (t >= start) & (t < end)
        t_seg = t[mask]
        x_v = np.array(x_vel).flatten()[mask]
        y_v = np.array(y_vel).flatten()[mask]
        N = len(t_seg)
        if N < 2:
            continue
        dt = np.mean(np.diff(t_seg))
        yf_x = np.fft.fft(x_v - np.mean(x_v))
        yf_y = np.fft.fft(y_v - np.mean(y_v))
        xf = np.fft.fftfreq(N, dt)
        idx = (xf > 0.01) & (xf < 1.0)
        if np.any(idx):
            if freq_common is None:
                freq_common = xf[idx]
            if np.array_equal(xf[idx], freq_common):
                fft_x_list.append(2.0/N * np.abs(yf_x[idx]))
                fft_y_list.append(2.0/N * np.abs(yf_y[idx]))
    if not fft_x_list or not fft_y_list:
        print('No valid intervals for FFT')
        return
    avg_fft_x = np.mean(np.vstack(fft_x_list), axis=0)
    avg_fft_y = np.mean(np.vstack(fft_y_list), axis=0)
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plt.plot(freq_common, avg_fft_x, label='Avg X Velocity Spectrum (0.01-1 Hz)', color='b')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.title('Average Low Frequency Spectrum of X Velocity (FFT)')
    plt.grid()
    plt.legend()
    plt.subplot(2, 1, 2)
    plt.plot(freq_common, avg_fft_y, label='Avg Y Velocity Spectrum (0.01-1 Hz)', color='r')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.title('Average Low Frequency Spectrum of Y Velocity (FFT)')
    plt.grid()
    plt.legend()
    plt.tight_layout()
```

# Calculate Mean X and Y Velocities for All Group 2 Files (y=0)
This cell computes the mean x and y velocities for each file of the form "Group2_x<number>_y0_r1.301.25.Vectrino Profiler.00000.mat" in the data folder and displays the results in a table.

```python
import glob
import pandas as pd

# Find all relevant files
mat_files = glob.glob(os.path.join('Filtered_Group2', 'Filtered_Group2', 'Group2_x*_y0_r1.301.25.Vectrino Profiler.00000.mat'))

results = []
for file in mat_files:
    data = scipy.io.loadmat(file)
    struct = data['Data'][0, 0]
    x_vel = struct['Profiles_VelX_Filt']
    y_vel = struct['Profiles_VelY_Filt']
    mean_x = np.mean(x_vel)
    mean_y = np.mean(y_vel)
    # Extract x number from filename
    x_number = int(file.split('_')[3][1:])
    results.append({'x_number': x_number, 'mean_x_velocity': mean_x, 'mean_y_velocity': mean_y})

# Create DataFrame and display
results_df = pd.DataFrame(results).sort_values('x_number', ascending=True)
results_df.reset_index(drop=True, inplace=True)
display(results_df)
```

```python
def furier_analysis_diff(t_time1, x_vel1, y_vel1, t_time2, x_vel2, y_vel2, name1, name2):
    # Define intervals
    intervals = [(0, 30), (30, 60), (60, 90), (90, 120)]
    fft_x_list_1, fft_y_list_1, freq_common_1 = [], [], None
    fft_x_list_2, fft_y_list_2, freq_common_2 = [], [], None
    for start, end in intervals:
        # First dataset
        t1 = np.array(t_time1).flatten()
        t2 = np.array(t_time2).flatten()
        # Find common time points in this interval
        mask1 = (t1 >= start) & (t1 < end)
        mask2 = (t2 >= start) & (t2 < end)
        t1_seg = t1[mask1]
        t2_seg = t2[mask2]
        x_v1 = np.array(x_vel1).flatten()[mask1]
        y_v1 = np.array(y_vel1).flatten()[mask1]
        x_v2 = np.array(x_vel2).flatten()[mask2]
        y_v2 = np.array(y_vel2).flatten()[mask2]
        # Match lengths by interpolating the longer to the shorter
        if len(t1_seg) > 1 and len(t2_seg) > 1:
            if len(t1_seg) > len(t2_seg):
                x_v1 = np.interp(t2_seg, t1_seg, x_v1)
                y_v1 = np.interp(t2_seg, t1_seg, y_v1)
                t_common = t2_seg
                x_v2 = x_v2
                y_v2 = y_v2
            elif len(t2_seg) > len(t1_seg):
                x_v2 = np.interp(t1_seg, t2_seg, x_v2)
                y_v2 = np.interp(t1_seg, t2_seg, y_v2)
                t_common = t1_seg
                x_v1 = x_v1
                y_v1 = y_v1
            else:
                t_common = t1_seg
            N = len(t_common)
            if N < 2:
                continue
            dt = np.mean(np.diff(t_common))
            yf_x1 = np.fft.fft(x_v1 - np.mean(x_v1))
            yf_y1 = np.fft.fft(y_v1 - np.mean(y_v1))
            yf_x2 = np.fft.fft(x_v2 - np.mean(x_v2))
            yf_y2 = np.fft.fft(y_v2 - np.mean(y_v2))
            xf = np.fft.fftfreq(N, dt)
            idx = (xf > 0.01) & (xf < 1.0)
            if np.any(idx):
                if freq_common_1 is None:
                    freq_common_1 = xf[idx]
                if np.array_equal(xf[idx], freq_common_1):
                    fft_x_list_1.append(2.0/N * np.abs(yf_x1[idx]))
                    fft_y_list_1.append(2.0/N * np.abs(yf_y1[idx]))
                    fft_x_list_2.append(2.0/N * np.abs(yf_x2[idx]))
                    fft_y_list_2.append(2.0/N * np.abs(yf_y2[idx]))
    if not fft_x_list_1 or not fft_x_list_2:
        print('No valid intervals for FFT')
        return
    avg_fft_x_1 = np.mean(np.vstack(fft_x_list_1), axis=0)
    avg_fft_y_1 = np.mean(np.vstack(fft_y_list_1), axis=0)
    avg_fft_x_2 = np.mean(np.vstack(fft_x_list_2), axis=0)
    avg_fft_y_2 = np.mean(np.vstack(fft_y_list_2), axis=0)
    diff_fft_x = avg_fft_x_1 - avg_fft_x_2
    diff_fft_y = avg_fft_y_1 - avg_fft_y_2
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plt.plot(freq_common_1, diff_fft_x, label=f'Diff X Velocity Spectrum ({name1} - {name2})', color='b')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude Difference')
    plt.title('Difference of Average X Velocity Spectra (FFT)')
    plt.grid()
    plt.legend()
    plt.subplot(2, 1, 2)
    plt.plot(freq_common_1, diff_fft_y, label=f'Diff Y Velocity Spectrum ({name1} - {name2})', color='r')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude Difference')
    plt.title('Difference of Average Y Velocity Spectra (FFT)')
    plt.grid()
    plt.legend()
    plt.tight_layout()
```

# Statistical Analysis of Amplitudes in Difference of Fourier Transforms (x12 - x497)
This cell extracts the amplitude values for each frequency in the difference of Fourier transforms, then calculates and reports the mean, median, mode, and standard deviation for both X and Y velocity spectra.

```python
from scipy import stats

def furier_diff_stats(t_time1, x_vel1, y_vel1, t_time2, x_vel2, y_vel2, name1, name2):
    intervals = [(0, 30), (30, 60), (60, 90), (90, 120)]
    fft_x_list_1, fft_y_list_1, freq_common_1 = [], [], None
    fft_x_list_2, fft_y_list_2, freq_common_2 = [], [], None
    for start, end in intervals:
        t1 = np.array(t_time1).flatten()
        t2 = np.array(t_time2).flatten()
        mask1 = (t1 >= start) & (t1 < end)
        mask2 = (t2 >= start) & (t2 < end)
        t1_seg = t1[mask1]
        t2_seg = t2[mask2]
        x_v1 = np.array(x_vel1).flatten()[mask1]
        y_v1 = np.array(y_vel1).flatten()[mask1]
        x_v2 = np.array(x_vel2).flatten()[mask2]
        y_v2 = np.array(y_vel2).flatten()[mask2]
        if len(t1_seg) > 1 and len(t2_seg) > 1:
            if len(t1_seg) > len(t2_seg):
                x_v1 = np.interp(t2_seg, t1_seg, x_v1)
                y_v1 = np.interp(t2_seg, t1_seg, y_v1)
                t_common = t2_seg
            elif len(t2_seg) > len(t1_seg):
                x_v2 = np.interp(t1_seg, t2_seg, x_v2)
                y_v2 = np.interp(t1_seg, t2_seg, y_v2)
                t_common = t1_seg
            else:
                t_common = t1_seg
            N = len(t_common)
            if N < 2:
                continue
            dt = np.mean(np.diff(t_common))
            yf_x1 = np.fft.fft(x_v1 - np.mean(x_v1))
            yf_y1 = np.fft.fft(y_v1 - np.mean(y_v1))
            yf_x2 = np.fft.fft(x_v2 - np.mean(x_v2))
            yf_y2 = np.fft.fft(y_v2 - np.mean(y_v2))
            xf = np.fft.fftfreq(N, dt)
            idx = (xf > 0.01) & (xf < 1.0)
            if np.any(idx):
                if freq_common_1 is None:
                    freq_common_1 = xf[idx]
                if np.array_equal(xf[idx], freq_common_1):
                    fft_x_list_1.append(2.0/N * np.abs(yf_x1[idx]))
                    fft_y_list_1.append(2.0/N * np.abs(yf_y1[idx]))
                    fft_x_list_2.append(2.0/N * np.abs(yf_x2[idx]))
                    fft_y_list_2.append(2.0/N * np.abs(yf_y2[idx]))
    if not fft_x_list_1 or not fft_x_list_2:
        print('No valid intervals for FFT')
        return
    avg_fft_x_1 = np.mean(np.vstack(fft_x_list_1), axis=0)
    avg_fft_y_1 = np.mean(np.vstack(fft_y_list_1), axis=0)
    avg_fft_x_2 = np.mean(np.vstack(fft_x_list_2), axis=0)
    avg_fft_y_2 = np.mean(np.vstack(fft_y_list_2), axis=0)
    diff_fft_x = avg_fft_x_1 - avg_fft_x_2
    diff_fft_y = avg_fft_y_1 - avg_fft_y_2
    # Calculate statistics
    stats_dict = {}
    for label, arr in zip(['X', 'Y'], [diff_fft_x, diff_fft_y]):
        stats_dict[label] = {
            'mean': np.mean(arr),
            'median': np.median(arr),
            'mode': float(stats.mode(arr, keepdims=True)[0]),
            'std': np.std(arr)
        }
    print('Statistics for amplitude differences (x12 - x497):')
    for label in ['X', 'Y']:
        print(f"{label} Velocity Spectrum:")
        print(f"  Mean: {stats_dict[label]['mean']:.6f}")
        print(f"  Median: {stats_dict[label]['median']:.6f}")
        print(f"  Mode: {stats_dict[label]['mode']:.6f}")
        print(f"  Std: {stats_dict[label]['std']:.6f}\n")
    return stats_dict, freq_common_1, diff_fft_x, diff_fft_y

# Run the analysis for x12 and x497
diff_stats, freqs, diff_fft_x, diff_fft_y = furier_diff_stats(t_time_12, x_vel_12, y_vel_12, t_time_497, x_vel_497, y_vel_497, 'x12', 'x497')
```

# Histogram of Amplitude Distributions for X and Y Velocity Spectra at x12 and x497

The following plots show the amplitude distributions (histograms) for the X and Y velocity spectra at locations x12 and x497, respectively. The frequency with the highest amplitude is annotated in each plot.

```python
def compute_avg_fft(t_time, x_vel, y_vel):
    intervals = [(0, 30), (30, 60), (60, 90), (90, 120)]
    fft_x_list, fft_y_list, freq_common = [], [], None
    for start, end in intervals:
        t = np.array(t_time).flatten()
        mask = (t >= start) & (t < end)
        t_seg = t[mask]
        x_v = np.array(x_vel).flatten()[mask]
        y_v = np.array(y_vel).flatten()[mask]
        N = len(t_seg)
        if N < 2:
            continue
        dt = np.mean(np.diff(t_seg))
        yf_x = np.fft.fft(x_v - np.mean(x_v))
        yf_y = np.fft.fft(y_v - np.mean(y_v))
        xf = np.fft.fftfreq(N, dt)
        idx = (xf > 0.01) & (xf < 1.0)
        if np.any(idx):
            if freq_common is None:
                freq_common = xf[idx]
            if np.array_equal(xf[idx], freq_common):
                fft_x_list.append(2.0/N * np.abs(yf_x[idx]))
                fft_y_list.append(2.0/N * np.abs(yf_y[idx]))
    if not fft_x_list or not fft_y_list:
        return None, None, None
    avg_fft_x = np.mean(np.vstack(fft_x_list), axis=0)
    avg_fft_y = np.mean(np.vstack(fft_y_list), axis=0)
    return freq_common, avg_fft_x, avg_fft_y
```

```python
import pandas as pd

x_values = (7, 12, 17, 22, 27, 32, 37, 42, 47, 52, 62, 72, 82, 92, 102, 112, 122, 132, 142, 152, 162, 172, 497)
results = []

for val in x_values:
    mat_file = os.path.join('Filtered_Group2', 'Filtered_Group2', f'Group2_x{val}_y0_r1.301.25.Vectrino Profiler.00000.mat')
    t_time, x_vel, y_vel = read_mat_file(mat_file)
    freqs, avg_fft_x, avg_fft_y = compute_avg_fft(t_time, x_vel, y_vel)
    if freqs is not None:
        # Y velocity only
        abs_y = np.abs(avg_fft_y)
        max_idx_y = np.argmax(abs_y)
        max_freq_y = freqs[max_idx_y]
        max_amp_y = abs_y[max_idx_y]
        mean_y = np.mean(abs_y)
        std_y = np.std(abs_y)
        stds_above_mean_y = (max_amp_y - mean_y) / std_y if std_y > 0 else np.nan
        results.append({
            'x_value': val,
            'Y_max_freq': max_freq_y,
            'Y_max_amp': max_amp_y,
            'Y_mean': mean_y,
            'Y_std': std_y,
            'Y_stds_above_mean': stds_above_mean_y
        })

results_df = pd.DataFrame(results)

print('Summary of Maximum Amplitudes, Frequencies, Mean, Std, and Std Above Mean for Each x_value (Y velocity only):')
display(results_df)
```
